- Para séries o valor é `YYY Season(s)`

Como faríamos para converter essa coluna em uma coluna numérica que informa a duração em minutos para os filmes?

**Importante**: É mais fácil se manipularmos apenas os filmes, portanto use o dataframe `filmes` neste exemplo.

- Aplique o método `.str.split(expand=True)` na coluna duration e observe o resultado
- Note que duas novas colunas foram criadas, uma com nome `0` e outra com nome `1`
- A informação que queremos está na coluna `0`, então elimine a outra usando `.drop()`
- Renomeie a coluna `0` para `duration_min` usando o método `.rename()` no dataframe

````python
df_filmes['duracao_min'] = (df_filmes['duracao']#cria a coluna duracao_min e adiciona os parametros
  .str.split(expand=True)#realiza o split de num e string
  .drop(columns=[1])#deleta a coluna
  .rename(columns={0: 'duracao_min'})#renomeia a coluna
)

df_filmes.columns #ver as colunas apos adicionar

# df_filmes['duracao_min'].isna().sum() #trazer quantidade de linhas nan na coluna
````
